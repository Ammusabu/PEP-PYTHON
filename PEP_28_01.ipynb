**INHERITANCE**

```python
class human:
    def eat(self):
        print("eating")
class Student(human):
    def study(self):
        print("studying")
s1=Student()
s1.study()
```

```python
class Animal:
    def eat(self):
        print("eating")
class Dog(Animal):
    def bark(self):
        print("barking")
s1=Dog()
s1.bark()
s1.eat()
```

```python
class Parent:
    def __init__(self):
        print("Parent Constructor Called")
class Child(Parent):
    pass
s1 = Child()
```

```python
class Nokia:
    def call(self):
        print("Nokia can call")
class iphone(Nokia):
    def internet(self):
        print("Iphone have internet")
        self.call()
s1 = iphone()
s1.internet()
```

```python
class Parent:
    def __init__(self):
        print("This is Parent Constructor")
class Child(Parent):
    def __init__(self):
        super().__init__()
        print("This is child Constructor")
s1=Child()
```

**Method Overriding**

```python
class Parent:
    def work(self):
        print("Parents are working")
class Child(Parent):
    def work(self):
        print("Childrens are playing")
s1=Child()
s1.work()
```

**MULTIPLE INHERITANCE**

```python
class Father:
    def Eyes(self):
        print("Child have father Eyes")
class Mother:
    def Nose(self):
        print("Child have mother Nose")
class Child(Father,Mother):
    def prop(self):
        self.Eyes()
        self.Nose()
        print("I am the child")
s1 = Child()
s1.prop()
```

```python
class Parent:
    def Shape(self):
        print("The Shape is Oval")
class Child(Parent):
    def Shape(self):
        super().Shape()
        print("The shape is circle")
s1 = Child()
s1.Shape()
```

```python
class Teacher:
    def teach(self):
        print("Teaching")
class Coder:
    def code(self):
        print("coding")
class Student(Teacher,Coder):
    pass
s1 = Student()
s1.teach()
s1.code()
```

```python
class parent:
    def __init__(self):
        self.__marks = 90 #private __

class child(parent):
    def show(self):
        print(self._parent__marks)

s1 = child()
s1.show()
```

```python
class parent:
    def __init__(self):
        self._marks = 90 #protected _

class child(parent):
    def show(self):
        print(self._marks)

s1 = child()
s1.show()

obj = parent()
print(obj._marks)
```

```python
class Account:
    def __init__(self,balance):
        self._balance = balance
    def showbalance(self):
        return self._balance

class SavingsAmount(Account):
    def addmoney(self, amount_to_add):
        self._balance += amount_to_add

    def showupdated(self):
        return self._balance

obj = SavingsAmount(500)
obj.addmoney(100)
print(obj.showupdated())
```
