```python
contact = {}

while True:
    print("\n---- Contact Book ----")
    print("1. Add contact")
    print("2. View contacts")
    print("3. Search contact")
    print("4. Delete contact")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    # 1. Add contact
    if choice == 1:
        name = input("Enter name: ")
        phone = input("Enter phone: ")
        contact[name] = phone
        print("Contact added successfully")

    # 2. View contacts
    elif choice == 2:
        if contact:
            print("\nSaved Contacts:")
            for name, phone in contact.items():
                print(name, ":", phone)
        else:
            print("No contacts found")

    # 3. Search contact
    elif choice == 3:
        name = input("Enter name to search: ")
        if name in contact:
            print(name, ":", contact[name])
        else:
            print("Contact not found")

    # 4. Delete contact
    elif choice == 4:
        name = input("Enter name to delete: ")
        if name in contact:
            del contact[name]
            print("Contact deleted successfully")
        else:
            print("Contact not found")

    # 5. Exit
    elif choice == 5:
        print("Exiting Contact Book")
        break

    else:
        print("Invalid choice! Please try again")
```

```python
Student = {}

while True:
    print("\n---- Student Database ----")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Delete Student")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    # 1. Add Student
    if choice == 1:
        name = input("Enter name: ")
        rollnumber = input("Enter roll number: ")
        course = input("Enter course: ")
        Student[name] = (rollnumber, course)
        print("Student added successfully")

    # 2. View Students
    elif choice == 2:
        if Student:
            print("\nSaved Students:")
            for name, details in Student.items():
                print("Name:", name)
                print("Roll No:", details[0])
                print("Course:", details[1])
                print("------------------")
        else:
            print("No students found")

    # 3. Search Student by Roll Number
    elif choice == 3:
        search_roll = input("Enter roll number to search: ")
        found = False
        for name, details in Student.items():
            if details[0] == search_roll:
                print("Student Found")
                print("Name:", name)
                print("Roll No:", details[0])
                print("Course:", details[1])
                found = True
                break
        if not found:
            print("Student not found")

    # 4. Delete Student
    elif choice == 4:
        name = input("Enter name to delete: ")
        if name in Student:
            del Student[name]
            print("Student deleted successfully")
        else:
            print("Student not found")

    # 5. Exit
    elif choice == 5:
        print("Exiting Student Database")
        break

    else:
        print("Invalid choice! Please try again")
```
