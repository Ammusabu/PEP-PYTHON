```python
#Methods
class Student:
    def __init__(self,name):
        self.name = name
    def hello(self):
        print("Hello",self.name)
    @staticmethod#only for class
    def college_name(): #
        print("LPU")
s1 = Student("Ammu")
s1.hello()
Student.college_name()
s1.college_name()
```

**Encapsulation**

```python
class Students:
    def __init__(self,marks):
        self.__marks = marks #private attribute
    def get_marks(self):
        return self.__marks
    def set_marks(self,newmarks):
        self.__marks = newmarks
s1 = Students(99)
s1.set_marks(100)
print(s1.get_marks())
```

```python
class Balance:
    def __init__(self,balance):
        self.__balance = balance
    def get_balance(self):
        return self.__balance
s1 = Balance(10000)
print(s1.get_balance())
```

```python
class Accounts:
    def __init__(self,balance):
        self.__balance = balance
    def get_balance(self):
        return self.__balance
    def deposite(self, amount):
        if amount > 0:
            self.__balance += amount
    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
s1 = Accounts(10000)
s1.deposite(1000)
s1.withdraw(500)
print(s1.get_balance())
```
