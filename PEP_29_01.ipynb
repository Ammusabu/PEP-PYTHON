2.   **ABSTRACT CLASS**



*   Only method name will be present
*   The method body will be absent
*   The child class will complete the method body











```python
class Payment:
    def pay(self,amount):
        pass
class Upi(Payment):
    def pay(self, amount):
        print("paid using upi:",amount)
class Card(Payment):
    def pay(self, amount):
        print("paid using card:",amount)
class cash(Payment):
    def pay(self, amount):
        print("paid using cash:",amount)
d1 = Upi()
d1.pay(90)
d2 = Card()
d2.pay(67)
d3 = cash()
d3.pay(78)
```

```python
class Shape:
    def start(self):
        print("Lets starts calculating Areas: ")
    def shape(self,area):
        pass

class circle(Shape):
    def shape(self,area):
        print("Area of circle:", 3.14 * area * area)

class Square(Shape):
    def shape(self,area):
        print("Area of square:", area * area)

class Rectangle(Shape):
    def shape(self,area):
        print("Rectangle value:", area)

circle_obj = circle()
circle_obj.start()
circle_obj.shape(3)

square_obj = Square()
square_obj.shape(4)

rectangle_obj = Rectangle()
rectangle_obj.shape(5)
```

```python
class Shape:
    def start(self):
        raise NotImplementedError("start() must be implemented")

    def area(self, value):
        raise NotImplementedError("area() must be implemented")
class Circle(Shape):
    def start(self):
        print("Starting area calculation")

    def area(self, value):
        print("Area of Circle:", 3.14 * value * value)


class Square(Shape):
    def start(self):
        print("Starting area calculation")

    def area(self, value):
        print("Area of Square:", value * value)
c = Circle()
c.start()
c.area(3)
c1 = Square()
c1.start()
c1.area(4)
```

```python
class Course:
    def info(self):
        print("This is a course")

    def duration(self):
        pass


class ExamInterface:
    def examtype(self):
        pass


class PythonCourse(Course, ExamInterface):

    def duration(self):
        print("Python course duration is 3 months")

    def examtype(self):
        print("Exam type is Online")


p = PythonCourse()
p.info()
p.duration()
p.examtype()
```

1. **COMPOSITION**
*  Use objects of each others class

*  Dont rely on inheritance
*  Introduced becuz to avoid error if the base class have any issues

```python
class Address:
    def __init__(self,city):
        self.city = city
    def show_address(self):
        print("City: ",self.city)
#Student Class
class Student:
    def __init__(self,name,city):
        self.name =name
        # composition
        #Creating obj of address calss inside stuent class
        self.address = Address(city)
    def show_student(self):
        print("Name: ",self.name)
        self.address.show_address()
s = Student("John","London")
s.show_student()
```

```python
class Engine:
    def start(self):
        print("Engine started")
class Car:
    def __init__(self):
        self.engine = Engine()
    def drive(self):

        self.engine.start()
        print("Car is driving")
s = Car()
s.drive()
```

```python
class Teacher:
    def SubjA(self):
        print("Teacher teaches subject A")
class Student:
    def __init__(self):
        self.teacher = Teacher()
    def study(self):
        self.teacher.SubjA()
        print("Student is studying A")

s = Student()
s.study()
```
