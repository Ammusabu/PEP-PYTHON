**REGEX PATTERN**

```python
import re
text = "Cat cot cut c@t"
result = re.findall("c.t",text)
print(result)
```

```python
#Start of string(^) and Ending of string($)
text = "Hello World"
print(bool(re.search("^Hello",text)))
print(bool(re.search("World$",text)))
```

```python
#0 or more(*)
text = "Hello"
result = re.findall("lo*",text)
print(result)
```

```python
#one or more(+)
text = "Hellooo"
result = re.findall("lo+",text)
print(result)
```

```python
#(?) Zero or one time
text = "Helloooo"
result = re.findall("lo?",text)
print(result)
```

```python
#([]) any one from given lsit
text = "HelJJ1loooo"
result = re.findall("[o]",text)
print(result)
```

```python
#[0-9] any digit
text = "HelJJ1loooo"
result = re.findall("[0-9]",text)
print(result)
```

```python
#([A-Z]) capital letter
text = "HelJJ1loooo"
result = re.findall("[A-Z]",text)
print(result)
```

```python
#([a-z]) small letter
text = "HelJJ1loooo"
result = re.findall("[a-z]",text)
print(result)
```

```python
#(A-Za-Z) both
text = "HelJJ1loooo"
result = re.findall("[A-Za-z]",text)
print(result)
```

```python
#(\d) digits
text = "HelJJ1loooo"
result = re.findall(r"\d",text)
print(result)
```

```python
#(\D) non digits
text = "HelJJ1loooo"
result = re.findall(r"\D",text)
print(result)
```

```python
#(\w) characters
text = "HelJJ1loooo"
result = re.findall(r"\w",text)
print(result)
```

```python
#(\W) non characters
text = "HelJ@J1loooo"
result = re.findall(r"\W",text)
print(result)
```

```python
#(\s) space
text = "HelJ@J 1loooo"
result = re.findall(r"\s",text)
print(result)
```

```python
#(\S) no space
text = "HelJ@J 1loooo"
result = re.findall(r"\S",text)
print(result)
```
