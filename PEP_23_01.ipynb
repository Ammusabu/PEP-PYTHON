```python
# String
s = "Python Programming"

print(s)                 # Prints the original string
print(len(s))            # Prints length of the string
print(s[0])              # Prints first character
print(s[-1])             # Prints last character
print(s[7:18])           # Prints substring using slicing

print(s.lower())         # Converts string to lowercase
print(s.upper())         # Converts string to uppercase

print(s)                 # Prints original string again
print(s.strip())         # Removes leading and trailing spaces

print(s.replace("Python", "Java"))  # Replaces a word in the string
print(s.split(" "))      # Splits string into a list of words

print(s)                 # Prints original string

print(s.find("P"))       # Finds position of character 'P'

word = s.split()         # Splits string into words
print(word)

b = "abc"
print(b.isdigit())       # Checks if string contains only digits
print(b.isalpha())       # Checks if string contains only alphabets

print(s.count("P"))      # Counts number of occurrences of 'P'
print(s.startswith("P")) # Checks if string starts with 'P'
print(s.endswith("g"))   # Checks if string ends with 'g'
print(s.capitalize())    #
#reverse a string
print(s[::-1])
```

```python
#palindrome check
s = "malayalam"
if s == s[::-1]:
    print("palindrome")
else:
    print("not palindrome")
```

```python
#make 2 pointer approach
s = "madam"

left = 0
right = len(s) - 1

while left < right:
    if s[left] != s[right]:
        print("Not a Palindrome")
        break
    left += 1
    right -= 1
else:
    print("Palindrome")

```

```python
#count characters of a string
m = "Apple"
for ch in m:
    print(ch,":", m.count(ch))
```

```python
#replace vowel with a star
l= "Lovely Professional University"
for ch in l:
    if ch in "aeiouAEIOU":
        l = l.replace(ch, "*")
    else:
        l = l.replace(ch, ch)
print(l)
```

```python
#same in another way
l = "Lovely Professional University"
result=""
for ch in l:
    if ch in "aeiouAEIOU":
        result += "*"
    else:
        result += ch
print(result)
#
```
