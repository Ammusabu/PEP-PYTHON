```python
#reverse a numer
x = __builtins__.int(input("Enter a number: "))

rev = 0

while x > 0:
    d = x % 10
    rev = rev * 10 + d
    x//= 10

print(rev)
```

```python
marks = [10,20,30,40,50]
marks.append(100)#will add the digit to end of the list
print(len(marks))
marks.insert(1,75)#will add that digit at index
print(marks)
marks.remove(30)#remove the digit i
print(marks)
marks.pop()#removes the last digit
print(marks)
print(marks[1:4])#from a to b index numbers
marks.sort()
print(marks)
marks.reverse()
print(marks)
print(max(marks))
print(min(marks))
```

```python
#Set Operations
numbers = {2,3,5,9,8,0}
print(numbers)
for i in numbers:
    print(i)
numbers.add(100)#only
print(numbers)
numbers.update([120,130])#multiple number additions
print(numbers)
numbers.remove(120)
numbers.discard(125)
print(numbers)
```

```python
a= {1,2,3,4,5}
b= {4,5,6,7,8}
print(a.union(b)) #a U b
print(a.intersection(b)) #a n b
print(a.difference(b))
```
