Design a console based "Employee Payroll System" using oops Python the system should calculate salary fordifferent types of employees using all oops concepts
Employee - Astract class
full time employee - child class
part time mployee - child class
salary - Encapsulated data
payroll system - contriller(Has -A)
output-
employee Created
salary: 50000
Employee created
salary: 40000

```python
class Employee:
    def __init__(self,emp_id,name):
        self.emp_id=emp_id
        self.name=name
    def calculate_salary(self):
        pass
class FullTimeEmployee(Employee):
    def __init__(self,emp_id,name,salary):
        super().__init__(emp_id,name)
        self._monthly_salary=salary
    def calculate_salary(self):
        return self._monthly_salary
class PartTimeEmployee(Employee):
    def __init__(self,emp_id,name,hours_worked,rate_per_hour):
        super().__init__(emp_id,name)
        self._hourly_worked=hours_worked
        self._rate_per_hour=rate_per_hour
    def calculate_salary(self):
        return self._hourly_worked * self._rate_per_hour
class PayrollApp:
    def __init__(self):
        self.employees=None
    def create_employee(self,employee):
        print("Employee Created")
        print("Salary:",employee.calculate_salary())
payroll = PayrollApp()

emp1 = FullTimeEmployee(101, "Ammu", 50000)
payroll.create_employee(emp1)

emp2 = PartTimeEmployee(102, "Anu", 80, 500)
payroll.create_employee(emp2)
```

Create a Library Management System where different library items calculate borrowing charges differently.
Library item (parent class)
Book and magazie (child class)
LibraryApp(main class)

Book IS-A libraryitem
Magazine IS-A libraryitem
LibraryApp HAS-A libraryitem

Output format
Item Type: Book
Borrow Days: 5
Borrowing Charge: 50
Or
Item Type: Magazine
Borrow Days: 3
Borrowing Charge: 30


```python
class LibraryItem:
    def __init__(self, borrow):
        self._borrow = borrow      #Encapsulation
    def cal_charge(self):
        pass                       #abstraction
class Book(LibraryItem):
    def cal_charge(self):           #Polimorphism
        return self._borrow * 10
class Magazine(LibraryItem):
    def cal_charge(self):
        return self._borrow * 10
class LibraryApp:
    def show_details(self, item, item_type): #HAS - A
        print("Item Type:", item_type)
        print("Borrow Days:", item._borrow)
        print("Borrowing Charge:", item.cal_charge())
app = LibraryApp()
book = Book(5)
app.show_details(book, "Book")
```
