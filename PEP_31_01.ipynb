**REGEX Pattern implementation**

```python
#Validate 10 digits number
import re

number = "1234572890"

if re.findall(r"\d{10}", number):
    print("Valid 10-digit number")
else:
    print("Invalid number")
```

```python
#extract the email from the string
import re
string = "Contact me at text@gmail.com and admin@yahoo.in"
emails = re.findall(r'[\w.-]+@[\w.-]+\.\w+', string)
print(emails)
```

```python
#extract number from a string
string = "Order: 123 Price: 45 Quantiy: 6"
print(re.findall(r'\d+', string))
print(re.findall(r"[0-9]+",string))
```

```python
#Validate a strong password(atleast 8 character,1 upper,1lower,1 special charact,1 digit )
import re
password = input()
if re.findall(r"[A-Z]",password) and re.findall(r"[a-z]",password) and re.findall(r"[0-9]",password) and re.findall(r"[^a-zA-Z0-9\s]",password) and len(password)>=8:
    print("Valid password")
else:
    print("Invalid password \nAdd atleast 8 characters\n1 upper case\n1 lowercase\n1 special character\n1 digit ")
```

```python
#validate a pancard
import re
pan_no = input()
if re.findall(r"[A-Z]{5}\d{4}[A-Z]",pan_no):
    print("Valid Pancard Number")
else:
    print("Invalid Pancard Number")
```

```python
import re

ip = input("Enter IP address: ")

pattern = r"^((25[0-5]|2[0-4][0-9]|1[0-9]{2}|[1-9]?[0-9])\.){3}(25[0-5]|2[0-4][0-9]|1[0-9]{2}|[1-9]?[0-9])$"

if re.match(pattern, ip):
    print("Valid IPv4 format")
else:
    print("Invalid IPv4 format")
```

```python
import random
import os

# Characters manually defined
lowercase = "abcdefghijklmnopqrstuvwxyz"
uppercase = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
digits = "0123456789"
special = "!@#$"

all_chars = lowercase + uppercase + digits + special


# 1. Generate password
def generate_password(length=10):
    password = ""
    for _ in range(length):
        password += random.choice(all_chars)
    return password


# 2. Count characters
def count_chars(password):
    counts = {
        "lowercase": 0,
        "uppercase": 0,
        "digits": 0,
        "special": 0
    }

    for ch in password:
        if ch in lowercase:
            counts["lowercase"] += 1
        elif ch in uppercase:
            counts["uppercase"] += 1
        elif ch in digits:
            counts["digits"] += 1
        elif ch in special:
            counts["special"] += 1

    return counts


# 3. Calculate strength score
def strength_score(password, counts):
    score = 0

    if len(password) >= 8:
        score += 2
    if counts["lowercase"] > 0:
        score += 2
    if counts["uppercase"] > 0:
        score += 2
    if counts["digits"] > 0:
        score += 2
    if counts["special"] > 0:
        score += 2

    return score


# 4. Label strength
def label(score):
    if score <= 4:
        return "Weak"
    elif score <= 8:
        return "Medium"
    else:
        return "Strong"


# ===== Main Program =====
password = generate_password()
counts = count_chars(password)
score = strength_score(password, counts)
strength = label(score)

# Save to file using os
os.makedirs("output", exist_ok=True)
path = os.path.join("output", "password.txt")

with open(path, "w") as f:
    f.write("Password: " + password + "\n")
    f.write("Score: " + str(score) + "/10\n")
    f.write("Strength: " + strength + "\n")
    f.write("Character Count: " + str(counts) + "\n")

# Output
print("Password:", password)
print("Score:", score)
print("Strength:", strength)
print("Counts:", counts)
print("Saved in:", path)
```
