**REGEX Pattern implementation**

```python
#Validate 10 digits number
import re

number = "1234572890"

if re.findall(r"\d{10}", number):
    print("Valid 10-digit number")
else:
    print("Invalid number")
```

```python
#extract the email from the string
import re
string = "Contact me at text@gmail.com and admin@yahoo.in"
emails = re.findall(r'[\w.-]+@[\w.-]+\.\w+', string)
print(emails)
```

```python
#extract number from a string
string = "Order: 123 Price: 45 Quantiy: 6"
print(re.findall(r'\d+', string))
print(re.findall(r"[0-9]+",string))
```

```python
#Validate a strong password(atleast 8 character,1 upper,1lower,1 special charact,1 digit )
import re
password = input()
if re.findall(r"[A-Z]",password) and re.findall(r"[a-z]",password) and re.findall(r"[0-9]",password) and re.findall(r"[^a-zA-Z0-9\s]",password) and len(password)>=8:
    print("Valid password")
else:
    print("Invalid password \nAdd atleast 8 characters\n1 upper case\n1 lowercase\n1 special character\n1 digit ")
```

```python
#validate a pancard
import re
pan_no = input()
if re.findall(r"[A-Z]{5}\d{4}[A-Z]",pan_no):
    print("Valid Pancard Number")
else:
    print("Invalid Pancard Number")
```

```python
import re

ip = input("Enter IP address: ")

pattern = r"^((25[0-5]|2[0-4][0-9]|1[0-9]{2}|[1-9]?[0-9])\.){3}(25[0-5]|2[0-4][0-9]|1[0-9]{2}|[1-9]?[0-9])$"

if re.match(pattern, ip):
    print("Valid IPv4 format")
else:
    print("Invalid IPv4 format")
```

```python
#hexadecimal
import re
hex_no = input()
if re.findall(r"^#?([A-Fa-f0-9]{6}|[A-Fa-f0-9]{3})$", hex_no):
    print("Valid Hexadecimal Number")
else:
    print("Invalid Hexadecimal Number")
```

```python
import random
import re
import os
import string

# Generate random password
def generate_password(n=10):
    chars = string.ascii_lowercase + string.ascii_uppercase + string.digits + "!@#$"
    return "".join(random.choice(chars) for _ in range(n))

# Calculate strength score
def strength_score(p):
    score = 0
    score += 2 if len(p) >= 8 else 0
    score += 2 if re.search(r"[a-z]", p) else 0
    score += 2 if re.search(r"[A-Z]", p) else 0
    score += 2 if re.search(r"\d", p) else 0
    score += 2 if re.search(r"[!@#$]", p) else 0
    return score

# Count character usage
def count_chars(p):
    return {
        "lowercase": len(re.findall(r"[a-z]", p)),
        "uppercase": len(re.findall(r"[A-Z]", p)),
        "digits": len(re.findall(r"\d", p)),
        "special": len(re.findall(r"[!@#$]", p))
    }

# Label strength
def label(score):
    if score <= 4:
        return "Weak"
    elif score <= 8:
        return "Medium"
    else:
        return "Strong"

# Main execution (NO user input)
password = generate_password()
score = strength_score(password)
level = label(score)
counts = count_chars(password)

# Save to file using os
folder = "output"
os.makedirs(folder, exist_ok=True)
path = os.path.join(folder, "password.txt")

with open(path, "w") as f:
    f.write(f"Password: {password}\n")
    f.write(f"Score: {score}/10\n")
    f.write(f"Strength: {level}\n")
    f.write(f"Character Count: {counts}\n")

# Output
print("Password:", password)
print("Score:", score)
print("Strength:", level)
print("Counts:", counts)
print("Saved in:", path)
```
